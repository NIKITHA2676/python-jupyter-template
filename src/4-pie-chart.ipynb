### Exercise 4: Pie Chart with Percentages

```python
try:
    import matplotlib.pyplot as plt
except ImportError as e:
    print(f"Error: {e}. Please ensure 'matplotlib' is installed.")
    exit()

# Data for the pie chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
colors = ['blue', 'red', 'green', 'orange']

try:
    # Validate input data
    if not all(isinstance(lst, list) for lst in [labels, sizes, colors]):
        raise TypeError("Error: Labels, sizes, and colors must all be lists.")

    if len(labels) != len(sizes) or len(labels) != len(colors):
        raise ValueError("Error: Labels, sizes, and colors must have the same length.")

    if any(not isinstance(size, (int, float)) or size < 0 for size in sizes):
        raise ValueError("Error: Sizes must be numeric (int or float) and non-negative.")

    if sum(sizes) != 100:
        raise ValueError("Error: Sizes must sum up to 100.")

    if any(not isinstance(label, str) or not label.strip() for label in labels):
        raise ValueError("Error: Labels must be non-empty strings.")

    if any(not isinstance(color, str) or not color.strip() for color in colors):
        raise ValueError("Error: Colors must be non-empty strings.")

    # Create pie chart
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors)

    # Add title
    plt.title("Pie Chart with Percentages")

    # Show the plot
    plt.show()

except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
